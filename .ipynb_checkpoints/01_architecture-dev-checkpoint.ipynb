Model Build

#### Paper Excpert:


The standard Transformer receives as input a 1D
sequence of token embeddings. To handle 2D images, we reshape the image $x ∈ \mathbb{R}^ {H×W×C}$ into a
sequence of flattened 2D patches:
$$
x_p ∈ \mathbb{R}
^{N×(P^{2}
·C)}
$$
Where: 
- $(H, W)$ is the resolution of the original
image
- $C$ is the number of channels
- $(P, P)$ is the resolution of each image patch
- $N = HW/P2$ is the resulting number of patches, which also serves as the effective input sequence length for the
Transformer.

The Transformer uses constant latent vector size $D$ through all of its layers, so we
flatten the patches and map to $D$ dimensions with a trainable linear projection (Eq. 1). We refer to
the output of this projection as the patch embeddings.

```python
from torch import nn

# Set Patch Size
patch_size = 16


# Patching Conv Layer
conv2d = nn.Conv2d(in_channels = 3,
                   out_channels = 768, # Embedding Size
                   kernel_size= patch_size,
                   stride = patch_size,
                   padding = 0)
conv2d
```

```python
image_out_of_conv = conv2d(image.unsqueeze(0)) # Add batch dimension to pass image through nn.layer
print(image_out_of_conv.shape)
```

```python
import random

random_idxs = random.sample(range(0,758), k=5)
print(f'Showing 5 random feature maps from idexes {random_idxs}')

# Create a Plot
fig, axs = plt.subplots(nrows= 1, ncols= 5, figsize=(12,12))

# Plot random feature map:
for i, idx in enumerate(random_idxs):
  image_conv_feature_map = image_out_of_conv[:,idx,:,:] # Index Feature maps of output tensor
  axs[i].imshow(image_conv_feature_map.squeeze().detach().numpy())
```

<div class="alert-info">
FLATTEN PATCHES

```python
# [BATCH, FEATURES, HEIGHT, WIDTH] - We want to flatten Height and Width dimensions

# Create flatten layer
flatten = nn.Flatten(start_dim=2, # flatten feature_map_height (start the flattening on the 2nd dimension)
                     end_dim=3) # flatten feature_map_width (dimension 3)
```

```python
# 1. View single image
plt.imshow(image.permute(1, 2, 0)) # adjust for matplotlib
plt.title(class_names[label])
plt.axis(False);
print(f"Original image shape: {image.shape}")

# 2. Turn image into feature maps
image_out_of_conv = conv2d(image.unsqueeze(0)) # add batch dimension to avoid shape errors
print(f"Image feature map shape: {image_out_of_conv.shape}")

# 3. Flatten the feature maps
image_out_of_conv_flattened = flatten(image_out_of_conv)
print(f"Flattened image feature map shape: {image_out_of_conv_flattened.shape}")
```

```python
# ** The paper ViT architecture wants [BATCH,PATCH,FEATURES] so we have to permute the flattened tensors

# Get flattened image patch embeddings in right shape 
image_out_of_conv_flattened_reshaped = image_out_of_conv_flattened.permute(0, 2, 1) # [batch_size, P^2•C, N] -> [batch_size, N, P^2•C]
print(f"Patch embedding sequence shape: {image_out_of_conv_flattened_reshaped.shape} -> [batch_size, num_patches, embedding_size]")
```

```python
class PatchEmbedding(nn.Module):
  def __init__(self,
               in_channels: int = 3,
               patch_size: int = 16,
               embedding_dim: int = 768):
    super().__init__()
    self.patcher = nn.Conv2d(in_channels = in_channels,
                             out_channels = embedding_dim,
                             kernel_size= patch_size,
                             stride = patch_size,
                             padding = 0)
    self.flatten = nn.Flatten(start_dim = 2,
                              end_dim = 3)
    self.patch_size = patch_size
    
  def forward(self, x):

    # Image size check
    image_resolution = x.shape[-1] # Check last dimension
    assert image_resolution % patch_size == 0, f'Input dims not evenly divisible by patch size, input size: {image_resolution}, patch size: {patch_size}'
    x = self.patcher(x)
    x = self.flatten(x)
    x = x.permute(0, 2, 1) # For desired input of next layer
    return x
```

```python
# Create an instance of patch embedding layer
patchify = PatchEmbedding(in_channels=3,
                          patch_size=16,
                          embedding_dim=768)

# Pass a single image through
print(f"Input image shape: {image.unsqueeze(0).shape}")
patch_embedded_image = patchify(image.unsqueeze(0)) # add an extra batch dimension on the 0th index, otherwise will error
print(f"Output patch embedding shape: {patch_embedded_image.shape}")
```

<div class="alert-info">
CLASS TOKENS

#### Paper Excerpt:

Similar to BERT’s `[class]` token, we prepend a learnable embedding to the sequence of embedded patches 
$(z^{0}_0 = xclass)$
, whose state at the output of the Transformer encoder $(z^{0}_L)$ serves as the
image representation $y$. Both during pre-training and fine-tuning, a classification head is attached to $(z^{0}_L)$. The classification head is implemented by a MLP with one hidden layer at pre-training
time and by a single linear layer at fine-tuning time.

Because there are only 3 classes to the toy dataset. The class token in only a single number

Also rememeber its only one classification per image so these class tokens are going to be inserted into the PATCH dimension. It will be a 768-long feature vector we tack on to the list of embedded patches

```python
# Get the batch size and embedding dimension
batch_size, embedding_dimension = patch_embedded_image.shape[0], patch_embedded_image.shape[-1]
# embedding_dimension = patch_embedded_image.shape[-1]

# Create the class token embedding as a learnable parameter that shares the same size as the embedding dimension (D)
class_token = nn.Parameter(torch.ones(batch_size, 1, embedding_dimension), # [batch_size, number_of_tokens, embedding_dimension]
                           requires_grad=True) # make sure the embedding is learnable

# Show the first 10 examples of the class_token
# print(class_token[:, :, :10])

# Print the class_token shape
print(f"Class token shape: {class_token.shape} -> [batch_size, number_of_tokens, embedding_dimension]")
print(f'Patch Embedded Image Shape: {patch_embedded_image.shape}')
```

```python
# Prepend Class Token
patch_embedded_image_with_class_embedding = torch.cat((class_token, patch_embedded_image), dim = 1) # Distribute these ones across the 2nd dimension

print(f"{patch_embedded_image_with_class_embedding.shape} - [BATCH,PATCHES,FEATURES]")
```

standard learnable 1D position embeddings, since we have not observed significant performance

sequence of embedding vectors serves as input to the encoder.

Class Tokens and Position Embeddings Assembly

```python
# 1. Patch Size
patch_size = 16 #Pixels

# 2. Print Dimensions of Input Image
print(f"Image Tensor Shape: {image.shape}")
height, width = image.shape[1], image.shape[2]

# 3. Get Image Tensor and Add Batch Dimension
x = image.unsqueeze(0) # Add batch dimension
print(f"Input Image Shape: {x.shape}")

# 4. Patch Layer Embedding:
patch_embedding_layer = PatchEmbedding(in_channels = 3,
                                       patch_size = patch_size,
                                       embedding_dim = 768)
# 5. Pass input through embedding layer
patch_embedding = patch_embedding_layer(x)
print(f"Patch Embedding Shape: {patch_embedding.shape}")

# 6. Create class token embedding
batch_size = patch_embedding.shape[0]
embedding_dimension = patch_embedding.shape[-1]
class_token = nn.Parameter(torch.ones(batch_size, 1, embedding_dimension),
                           requires_grad = True)
print(f"Class Token embedding shape: {class_token.shape}")

# 7. Prepend class token embedding to patch embedding
patch_embedding_class_token = torch.cat((class_token, patch_embedding), dim=1)
print(f"Patch Embedding with class token shape: {patch_embedding_class_token.shape}")

# 8. Create Position Embedding 
number_of_patches = int((height * width)) / patch_size**2
position_embedding = nn.Parameter(torch.ones(1, int(number_of_patches+1), embedding_dimension),
                                  requires_grad = True)

# 9. Add Position Embedding to Patch Embedding with Class Token
patch_and_position_embedding = patch_embedding_class_token + position_embedding
print(f"Patch and Postition Embedding Shape: {patch_and_position_embedding.shape}")
```

Transformer Encoder Layer

<div class="alert-info">
MULTI-HEAD SELF ATTENTION (MSA) BLOCK

```python
class MultiHeadSelfAttentionBlock(nn.Module):
  """ Creates MSA Block"""

  def __init__(self,
               embedding_dim:int = 768, # Features in incoming vectors
               num_heads:int = 12, # Number of 'feature attention maps'
               attn_dropout:int=0): # Paper doesn't impliment dropout in their attention layers
    super().__init__()
    
    # Create Norm Layer
    self.layer_norm = nn.LayerNorm(normalized_shape=embedding_dim)

    # Create Multihead Attention Layer
    self.multihead_attn = nn.MultiheadAttention(embed_dim = embedding_dim,
                                                num_heads = num_heads,
                                                dropout = attn_dropout,
                                                batch_first = True) # Tells model to expect input of the form [BATCH, PATCHES, FEATURES]

  def forward(self,x):
    x = self.layer_norm(x)
    attn_output, _ = self.multihead_attn(query = x,
                                      key = x,
                                      value = x, 
                                      need_weights = False) # Optional parameter of returning weights along with activations
    return attn_output
```

```python
# Instantiate MSA block
multihead_self_attention_block = MultiHeadSelfAttentionBlock(embedding_dim = 768,
                                                             num_heads = 12)

# Pass through test input
patched_image_through_msa_block = multihead_self_attention_block(patch_and_position_embedding)

print(f"Input shape of MSA Block: {patch_and_position_embedding.shape}")
print(f"Output Shape of MSA Block: {patched_image_through_msa_block.shape}")
```

<div class="alert-info">
FULLY CONNECTED (MLP) LAYER

```python
class MLPBlock(nn.Module):
  def __init__(self,
               embedding_dim:int=768, # Input size
               mlp_size:int = 3072, # Fully connected layer size
               dropout:int = 0.1):
    super().__init__()

    # Norm Layer
    self.layer_norm = nn.LayerNorm(normalized_shape = embedding_dim)

    # Create the MLP
    self.mlp = nn.Sequential(
        nn.Linear(in_features = embedding_dim,
                  out_features = mlp_size),
        nn.GELU(),
        nn.Dropout(p = dropout),
        nn.Linear(in_features = mlp_size,
                  out_features = embedding_dim),
        nn.Dropout(p = dropout)
    )

  def forward(self,x):
    x = self.layer_norm(x)
    x = self.mlp(x)
    return x
```

```python
# instantiate MLP Block
mlp_block = MLPBlock(embedding_dim = 768,
                     mlp_size = 3072,
                     dropout= 0.1)

patched_image_through_mlp_block = mlp_block(patched_image_through_msa_block)
print(f"MLP Block Input Shape: {patched_image_through_msa_block.shape}")
print(f"MLP Block Input Shape: {patched_image_through_mlp_block.shape}")
```

<div class="alert-info">
MSA + MLP COMPOSITION

```python
class TransformerEncoderBlock(nn.Module):
  def __init__(self,
               embedding_dim: int = 768,
               num_heads:int = 12,
               mlp_size:int = 3072,
               mlp_dropout:int = 0.1,
               attn_dropout:int = 0
               ):
    
    super().__init__()

    # MSA Block
    self.msablock = MultiHeadSelfAttentionBlock(
        embedding_dim = embedding_dim,
        num_heads = num_heads,
        attn_dropout = attn_dropout)
    
    # MLP Block
    self.mlpblock = MLPBlock(
        embedding_dim = embedding_dim,
        mlp_size = mlp_size,
        dropout = mlp_dropout
    )

  def forward(self,x):
    x = self.msablock(x) + x # MSA plus skip connection
    x = self.mlpblock(x) + x # MLP plus skip connection
    return x
```

<div style="background-color:moccasin; color:black; font-size:20px">
Full Assembly

```python
class ViT(nn.Module):
  def __init__(self,
               img_size: int = 224, # Input Dim
               in_channels = 3, # Color Channels
               patch_size:int = 16, # Size of patches to be cut out
               num_transformer_layers: int = 12, # num transformer blocks
               embedding_dim:int = 768, # Size of patch feature vectors
               mlp_size: int = 3072, # Fully Connected size
               num_heads:int = 12, # Num attention 'feature maps'
               attn_dropout:int = 0, # None implimented in paper
               mlp_dropout:int = .1, # Dropout layers in mlp block
               embedding_dropout:int = 0.1, # Dropout layer after position + class token embeddings
               num_classes: int = 1000):
    
    super().__init__()

    assert img_size % patch_size == 0, f"Image size ({img_size})not cleanly divisible by patch size({patch_size})"

    # Patch Size
    self.num_patches = (img_size * img_size) // patch_size**2

    # Class Embedding 
    self.class_embedding = nn.Parameter(data = torch.randn(1,1,embedding_dim),
                                        requires_grad = True)
    # Position Embedding
    self.position_embedding = nn.Parameter(data = torch.randn(1, self.num_patches +1, embedding_dim))

    # Embedding Dropout Layer
    self.embedding_dropout = nn.Dropout(p=embedding_dropout)

    # Patch Embedding Layer
    self.patch_embedding = PatchEmbedding(in_channels = in_channels,
                                          patch_size = patch_size,
                                          embedding_dim = embedding_dim)
    # Transformer Encoder Block
    self.transformer_encoder = nn.Sequential(*[TransformerEncoderBlock(embedding_dim = embedding_dim, # Use '* syntax' to make (num_transformer_layers = 12) transformer blocks
                                                                       num_heads = num_heads,         # I guess this means you can't reuse a module in the forward because it would probably update the
                                                                       mlp_size = mlp_size,            # same layer when it should be updating another identical layer
                                                                       mlp_dropout = mlp_dropout) for _ in range(num_transformer_layers)])
    # Classifier Head
    self.classifier = nn.Sequential(
        nn.LayerNorm(normalized_shape = embedding_dim),
        nn.Linear(in_features = embedding_dim,
                  out_features = num_classes)
    )

  def forward(self,x):

    batch_size = x.shape[0] # Get batch size

    class_token = self.class_embedding.expand(batch_size, -1, -1) # Flexibly allows token to accommodate the shape of an inputted batch size ('-1' means to infer dimensions)

    # Create Patch Embedding
    x = self.patch_embedding(x)

    # Class Token
    x = torch.cat((class_token, x), dim=1)

    # Position Embedding
    x = self.position_embedding + x

    # Embedding Dropout
    x = self.embedding_dropout(x)

    # Transformer-Encoder Block
    x = self.transformer_encoder(x)

    # Zero-ith logit (class token) to classifier
    x = self.classifier(x[:,0])

    return x
```
